```python
from langchain_community.retrievers import WikipediaRetriever

def main():
    # Step 1: Initialize retriever
    retriever = WikipediaRetriever(
        top_k_results=2,
        lang="en"
    )

    # Step 2: Query
    query = "Agentic ai"

    # Step 3: Fetch documents
    docs = retriever.invoke(query)

    # Step 4: Debug check
    if not docs:
        print("No results found. Try a different query.")
        return

    print(f"Total documents fetched: {len(docs)}\n")

    # Step 5: Display results properly
    for i, doc in enumerate(docs, 1):
        print(f"--- Document {i} ---")
        print("Title:", doc.metadata.get("title", "N/A"))
        print("Source:", doc.metadata.get("source", "N/A"))
        print("\nContent Preview:\n")
        print(doc.page_content[:500])  # first 500 chars
        print("\n" + "="*50 + "\n")


if __name__ == "__main__":
    main()
```
